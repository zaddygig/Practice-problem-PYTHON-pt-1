**9. Write a Python program to convert temperatures to and from Celsius,
Fahrenheit. [ Formula: c = (f-32)(5/9)]**

```python
temp = float(input("Enter Temparature"))
unit = input("enter unit(C or F)    :")

if unit == 'C' or unit == 'c':
  newt= 9 / 5 *  temp  + 32
  print("temparatur in Fahranheit =",newt)
elif unit == 'F' or unit == 'f':
    newt = 5 / 9 * (temp - 32)
    print("temparature in celsius=",newt)
else :
  print ("unknown unit")
    
```

**10. Write a Python program to construct the following pattern, using a nested for loop**

```python
n=7;
for i in range(n):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```
